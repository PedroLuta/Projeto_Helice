```python
%matplotlib inline

from mosalib.mosa import Anneal
from pandas import read_csv
```

O que queremos aqui é obter uma liga binária com componentes fictícios, onde o objetivo do otimizador é maximizar uma propriedade X qualquer (também fictícia) e minimizar o custo da liga (que também é fictício).

Os dados utilizados foram gerados utilizando o script datagen.py e... são fictícios.

Esses dados são carregados usando o Pandas a partir do arquivo "dataset", em formato CSV.

É preciso definir uma população inicial, que é a partir de onde serão obtidos os valores que irão compôr as soluções. No caso em tela, temos dois grupos nessa população. 

O primeiro é o grupo "Component", obtido do arquivo "dataset", que é uma lista de valores discretos, onde cada elemento da liga binária é listado. Abra o arquivo "dataset" para entender melhor.

O segundo é o grupo "Concentration", que, por ser uma lista vazia, implica que é um contínuo entre dois limites definidos mais abaixo.

As soluções irão conter os mesmos grupos, "Component" e "Concentration". Em "Component", teremos os dois componentes da liga binária. Em "Concentration", a porcentagem do segundo componente.

```python
df=read_csv("dataset")
population={"Component":df.Component.values.tolist(),
            "Concentration":[]}
```

A função a ser otimizada é definida abaixo. 

Ela recebe uma solução proposta para o problema como argumento.

Ela retorna uma lista. No caso deste problema, a lista contém o valor da propriedade X (multiplicado por -1, porque queremos maximizar X) e o custo da liga binária para a solução proposta.

```python
def func(solution):
    elem1=df[df.Component==solution["Component"][0]]
    elem2=df[df.Component==solution["Component"][1]]
    conc=solution["Concentration"][0]
    xmax=df.X.values.max()
    xmin=df.X.values.min()
    costmax=df.Cost.values.max()
    costmin=df.Cost.values.min()
    x1=elem1.X.values[0]
    cost1=elem1.Cost.values[0]
    x2=elem2.X.values[0]
    cost2=elem2.Cost.values[0]  
    x=(x1*(1.0-conc)+x2*conc)/(xmax-xmin)
    cost=(cost1*(1.0-conc)+cost2*conc)/(costmax-costmin)
        
    return [-x,cost]
```

Um objeto Anneal é instanciado, algumas configurações da otimização são definidas, assim como a população que será considerada na otimização (veja acima).

```python
opt=Anneal()
opt.setarchiveconfig(archivesize=100,maxarchivereject=1000)
opt.settemp(initemp=1.0,niter=100,ntemp=50)
opt.setxconfig(xnel={"Component":2,
                     "Concentration":1},
               selweight={"Component":1.0,
                          "Concentration":4.0},
               xdistinct={"Component":True},
               exchangeprob={"Component":1.0,
                             "Concentration":1.0},
               xstep={"Component":5,
                      "Concentration":0.05},
               xbounds={"Concentration":[0.0,0.1]})
opt.setpopulation(population)
```

Na linha abaixo, a otimização realmente acontece.

```python
opt.evolve(func)
```

Imprime o conteúdo do Archive e plota X vs Cost

```python
opt.printx()
opt.plotfront()
```

Expurga do Archive as soluções dominadas, imprime as não dominadas e plota X vs Cost

```python
purged=opt.purgedominated()
opt.printx(purged)
opt.plotfront(purged)
```

Seleciona apenas as soluções que satisfaçam o threshold

```python
trimmed=opt.trimx(xset=purged,thresholds=[-1.5,0.5])
opt.printx(trimmed)
```

Seleciona as dez soluções com menor custo

```python
sol1=opt.reducex(xset=trimmed,index=1,nel=10)
opt.printx(sol1)
```

Das soluções com menor custo, seleciona as cinco com os maiores valores (multiplicados por -1, não custa lembrar) da propriedade X

```python
sol2=opt.reducex(xset=sol1,index=0,nel=5)
opt.printx(sol2)
```

E é isso. Em princípio, ao final do processo, você, o ser humano, tem um pequeno subconjunto de ligas binárias com valores aceitáveis da propriedade X e a um custo que está disposto a pagar.
